```python
#Import Dependencies
import os
import pandas as pd
```

```python
#Write path to School Data CSV folder that was created from CSV creator notebook.
path = "./Resources/DataBySchoolCSVs"
csvList = os.listdir(path)
#Only 2008 - 2015 Data
csvList = csvList[12:]
```

```python
#Create list of keys which is basically just the year and the quarter
keys = []
for i in range(len(csvList)):
    keys.append(csvList[i][:-4])
```

```python
#Create a dictionary of dataframes where each key written above corresponds to the df for each csv file
#Each application cycle contains six quarters of data.
d = {}
for i in range(len(csvList)):
    d[str(keys[i])] = pd.read_csv(f'{path}/{csvList[i]}')[['OPE ID','School','State','Zip Code','Quarterly Total','Award Year To Date Total']]
```

```python
#Create dataframes for a single application cycle. Begin by populating only the static data for each
df0809 = d['20082009Qtr1'][['OPE ID','School','State','Zip Code']]
df0910 = d['20092010Qtr1'][['OPE ID','School','State','Zip Code']] 
df1011 = d['20102011Qtr1'][['OPE ID','School','State','Zip Code']]
df1112 = d['20112012Qtr1'][['OPE ID','School','State','Zip Code']]
df1213 = d['20122013Qtr1'][['OPE ID','School','State','Zip Code']]
df1314 = d['20132014Qtr1'][['OPE ID','School','State','Zip Code']]
df1415 = d['20142015Qtr1'][['OPE ID','School','State','Zip Code']]
df1516 = d['20152016Qtr1'][['OPE ID','School','State','Zip Code']]
```

```python
#Iterate across all 6 quarters of data in order to populate quartery totals and year to date awards.
for i in range(6):
    df0809[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20082009Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
    df0910[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20092010Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
    df1011[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20102011Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
    df1112[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20112012Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
    df1213[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20122013Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
    df1314[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20132014Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
    df1415[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20142015Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
    df1516[[f'Q{i+1} Total',f'YTD Q{i+1}']] = d[f'20152016Qtr{i+1}'][['Quarterly Total','Award Year To Date Total']]
```
